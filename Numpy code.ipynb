```python
import numpy as np
import time

# Create big data
list_data = list(range(10_000_000))
array_data = np.arange(10_000_000)

# List timing
start = time.time()
list_squared = [x**2 for x in list_data]
end = time.time()
print(f"List took: {end - start:.5f} seconds")

# NumPy timing
start = time.time()
array_squared = array_data ** 2
end = time.time()
print(f"NumPy took: {end - start:.5f} seconds")
```

```python
#Operations on np array
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)   # [5 7 9]
print(a - b)   # [-3 -3 -3]
print(a * b)   # [4 10 18]
print(a / b)   # [0.25 0.4 0.5]
print(a + 10)   # [11 12 13]
print(a * 2)    # [2 4 6]
print(a > 2)  # [False False  True]
print(b == 5) # [False  True False]
print(a.sum())    # 6
print(b.min())    # 4
print(b.max())    # 6
print(a.mean())   # 2.0
print(np.median(b)) # 5.0
```

```python
#Dot product
import numpy as np
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(3,2)
print(a.dot(b))
# 2x3 and 3x2
# 2x3 and 2x3 Not correct
# according to the matrix multiplication rule
```

```python
import numpy as np
a = np.arange(12).reshape(3,4)
print(a)
print(a.min(axis=0)) # min among all the coloums
print(a.min(axis=1)) # min among all the rows
print(a.mean(axis=0))
print(np.median(a))
print(np.sin(a))
print(np.exp(a))
```

```python
import numpy as np
a = np.arange(6,12).reshape(2,3)
print(a.ravel() ) # convert all the 2,3,4D arrays to the 1 d array
a.transpose()
b= np.arange(6).reshape(2,3)
print(np.hstack((a,b))) # join two arrays in  horizontal
print(np.vstack((a,b)))  # join two arrays in  vertical

#Splitting
```

```python
import numpy as np

a = np.arange(9).reshape(3, 3)
# Splitting into 3 equal parts along columns (axis=1)
print(np.split(a, 3, axis=1))

a = np.arange(12).reshape(2, 6)
print(np.hsplit(a, 3))  # splits into 3 parts column-wise

a = np.arange(12).reshape(4, 3)
print(np.vsplit(a, 2))  # splits into 2 parts row-wise
```
